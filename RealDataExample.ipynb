```python
# load in Ukraine Russia data
with open('full_hashtag_data_sentiments_matched_bot_class.pkl','rb') as f:
    df = pickle.load(f)
```

```python
# sort by mean bot classification probability and then count
d_temp = df[~df.BotClassificationP.isna()].groupby('author_id').agg({'BotClassificationP':['mean','count']})

# filter to get counts only larger than 200
d_temp = d_temp[d_temp[('BotClassificationP','count')] > 200]
d_temp.sort_values([('BotClassificationP','mean'), ('BotClassificationP','count')],ascending=[True, False])

# df.groupby('author_id').agg({'BotClassificationP':['mean','count']}).sort_values(('BotClassificationP','mean'),ascending=False)
```
