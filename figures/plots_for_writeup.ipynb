# plot showing av entropy is constant vs possible entropy rates

```python
x = np.arange(0,10,0.1)
y = 0.5*np.ones(100)

# sinusoidal
y1 = (np.sin(6*x/(np.pi))+5)/10

# linear inc
y2 = (x/2 + 5/2)/10

# step
y3 = [.35 if i<5 else .65 for i in x]
```

```python
sns.lineplot(x=x,y=y, linewidth=5, color='black')
sns.lineplot(x=x,y=y1, linewidth=2, linestyle='dashed')
sns.lineplot(x=x,y=y2, linewidth=2, linestyle='dashed')
sns.lineplot(x=x,y=y3, linewidth=2, linestyle='dashed')
plt.ylabel('Entropy rate')
plt.xlabel('Time')
plt.savefig('figures/average_entropy.pdf')
```
